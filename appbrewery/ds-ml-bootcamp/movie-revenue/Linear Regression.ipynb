```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Importing Box Office cost revenue data

```python
data = pd.read_csv("cost_revenue_clean.csv")
data = data.apply(pd.to_numeric, errors = 'coerce')
```

```python
data.describe()
```

```python
prod_budget = pd.DataFrame(data, columns = ["production_budget_usd"])
revenue = pd.DataFrame(data, columns = ["worldwide_gross_usd"])
```

alpha = transparency

```python
plt.figure(figsize = (10, 6))
plt.scatter(prod_budget, revenue, alpha = 0.3)
plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget US$")
plt.ylabel("Worldwide Gross US$")
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(prod_budget, revenue)
```

Show Coefficient $\theta_1$:

```python
regression.coef_
```

Intercept:

```python
regression.intercept_
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(prod_budget, revenue, alpha = 0.3)
plt.plot(prod_budget, regression.predict(prod_budget), color='red', linewidth = 4)

plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget US$")
plt.ylabel("Worldwide Gross US$")
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

Getting $R^2$ from Regression

```python
regression.score(prod_budget, revenue)
```
